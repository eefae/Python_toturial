````python
import os
from matplotlib import pyplot as plt
from keras.src.datasets import mnist

os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

action = mnist
(train_images, train_labels), (test_images, test_labels) = action.load_data()

def show_image(img,img_label,imglist):
    gif = plt.gcf()
    gif.set_size_inches(10, 10)
    ai = 1
    for i in imglist:
        ax = plt.subplot(5,5,ai)
        ai += 1
        ax.set_title("label:" + str(img_label[i]),fontsize=10)
        ax.imshow(img[i], cmap='binary')
        ax.set_xticks([])
        ax.set_yticks([])
    plt.show()

if __name__ == "__main__":
    image_list = [x for x in range(0,10)]
    show_image(train_images, train_labels, image_list)
````
